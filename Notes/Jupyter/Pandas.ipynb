```python
# import (larger) DataFrame from .csv file
movies_df = pd.read_csv('data/IMDB-Movie-Data.csv', index_col='Title')
# display head of DataSet (default: 5)
movies_df.head()
```

```python
# combine datasets (appending to self)
temp_df = movies_df.append(movies_df)
temp_df.head()
```

```python
# remove duplicate entries
temp_df.drop_duplicates(inplace=True)
temp_df.head()
```

```python
# list columns in DataFrame
movies_df.columns
```

```python
# rename columns in DataFrame
movies_df.rename(columns={
	'Runtime (Minutes)': 'Runtime',
	'Revenue (Millions)': 'Revenue_millions'
}, inplace=True)
movies_df.head()
```

```python
# rename all columns to lowercase
movies_df.columns = [col.lower() for col in movies_df]
movies_df.head()
```

```python
# show sum of incomplete data
movies_df.isnull().sum()
```

```python
# remove NaN data, display new shape
movies_df.dropna(axis=1).shape
```

```python
# display single column
revenue=movies_df['revenue_millions']
revenue.head()
```

```python
# find mean of column
revenue_mean=revenue.mean()
revenue_mean
```

```python
# fill NaN data with mean data
revenue.fillna(revenue_mean, inplace=True)
revenue.head()
```

```python
# return result of string condition
(movies_df['director'] == 'Ridley Scott').head(10)
```

```python
# return data based on string condition
movies_df[movies_df['director'] == 'Ridley Scott'].head()
```

```python
# return data based on numerical condition
movies_df[movies_df['rating'] >= 8.6].head()
```

```python
# return data based on multiple conditions (note: vertical bar means "or")
movies_df[
	(movies_df['director'] == 'Christopher Nolan') |
	(movies_df['director'] == 'Ridley Scott')
].head()
```

```python
# return data based on multiple conditions (returns same as above, but code is more concise)
movies_df[movies_df['director'].isin(
	['Christopher Nolan', 'Ridley Scott']
)].head()
```

```python
# return data based on many conditions (note: ampersand means "and")
movies_df[
	(
		(movies_df['year'] >= 2005) &
		(movies_df['year'] <= 2010) &
		(movies_df['rating'] > 8.0) &
		(movies_df['revenue_millions'] < movies_df['revenue_millions'].quantile(0.25))
	)
]
```

```python
# create function to iterate through data
def rating_function(x):
	if x >= 8.0:
		return 'good'
	else:
		return 'bad'

# use function on dataset
movies_df['rating_category'] = movies_df['rating'].apply(rating_function)
movies_df.head()
```

```python
# create scatter plot from DataFrame
movies_df.plot(
	kind='scatter',
	x='rating',
	y='revenue_millions',
	title='Revenue (millions) v Rating'
);
```

```python
# show linear regression
fig = plt.figure(figsize = (10, 7))
sns.regplot(
	x = movies_df.rating,
	y = movies_df.revenue_millions
);
plt.title('Rating v Revenue');
```

```python
# plot histogram based on column
movies_df['rating'].plot(
	kind='hist',
	title='Rating'
);
```

```python
# plot boxplot based on column
movies_df['rating'].plot(kind='box');
```

```python
# plot boxplot based on numerical and categorical data
movies_df.boxplot(
	column='revenue_millions',
	by='rating_category'
);
```
